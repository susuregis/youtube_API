```python
def get_channel_views(channel_username):
    request = youtube.channels().list(
        part="statistics",
        forUsername=channel_username
    )
    response = request.execute()

    total_views = int(response["items"][0]["statistics"].get("viewCount", 0))
    return total_views

def get_highlight_videos(channel_id, year):
    start_date = f"{year}-01-01T00:00:00Z"
    end_date = f"{year + 1}-01-01T00:00:00Z"

    request = youtube.search().list(
        part="snippet",
        channelId=channel_id,
        publishedAfter=start_date,
        publishedBefore=end_date,
        q="highlights",
        type="video",
        maxResults=25
    )
    response = request.execute()
    return response['items']

def get_video_views(video_id):
    """ObtÃ©m o nÃºmero de visualizaÃ§Ãµes de um vÃ­deo"""
    request = youtube.videos().list(
        part="statistics",
        id=video_id
    )
    response = request.execute()

    return int(response["items"][0]["statistics"].get("viewCount", 0))

def main():
    channel_username = "Formula1"

    total_channel_views = get_channel_views(channel_username)
    print(f"\ Total de visualizaÃ§Ãµes do canal: {total_channel_views}\n")
    print("-" * 80)


    request = youtube.channels().list(part="id", forUsername=channel_username)
    response = request.execute()
    channel_id = response["items"][0]["id"]


    highlight_videos = get_highlight_videos(channel_id, 2024)

    video_data = []

    for video in highlight_videos:
        video_id = video["id"]["videoId"]
        title = video["snippet"]["title"]
        views = get_video_views(video_id)

        percentage = (views / total_channel_views) * 100 if total_channel_views > 0 else 0

        video_data.append((title, views, percentage, video_id))


    video_data.sort(key=lambda x: x[1], reverse=True)


    for title, views, percentage, video_id in video_data:
        print(f" TÃ­tulo: {title}")
        print(f"VisualizaÃ§Ãµes: {views}")
        print(f" Representa {percentage:.4f}% do total de visualizaÃ§Ãµes do canal")
        print(f" Link: https://www.youtube.com/watch?v={video_id}")
        print("-" * 80)

if __name__ == "__main__":
    main()
```
